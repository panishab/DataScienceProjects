```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPRegressor
from dmba import classificationSummary, regressionSummary
```

```python
df_Toyota = pd.read_csv(r"C:\Users\anish\Downloads\ToyotaCorolla.csv")
```

```python

# select predictors
choosed = ['Price','Age_08_04', 'KM','Fuel_Type', 'HP', 'Automatic', 'Doors', 'Quarterly_Tax', 'Mfr_Guarantee', 'Guarantee_Period', 'Airco', 'Automatic_airco', 'CD_Player', 'Powered_Windows', 'Sport_Model', 'Tow_Bar']
df_Toyota=df_Toyota[choosed]
df_Toyota.head()
```

```python
# create dummies for categorical predictors

df_Toyota= pd.get_dummies(df_Toyota,columns=['Fuel_Type'],drop_first=True)
df_Toyota

```

```python
# scale the predictors and outcome variable
scaler = MinMaxScaler()
df_scaled= scaler.fit_transform(df_Toyota)
```

```python
#create dataframe of scaled 
df_scaled=pd.DataFrame(df_scaled, columns=df_Toyota.columns)
```

```python
df_scaled.head()
```

```python
#split in targets as y and features as X
X=df_scaled.drop(columns='Price')
y=df_scaled['Price']
```

```python
#split the data in training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.4,random_state=1)
```

```python
mlp=MLPRegressor(hidden_layer_sizes=(2),activation='logistic', solver='lbfgs', random_state=1)
mlp.fit(X_train,y_train)
```

```python
#evaluate model in traing set
y_train_pred=mlp.predict(X_train)
train_rmse=mean_squared_error(y_train,y_train_pred,squared=False)
print("Training RMSE:", train_rmse)
```

```python
#evaluate model in traing set
y_test_pred=mlp.predict(X_test)
test_rmse=mean_squared_error(y_test,y_test_pred, squared=False)
print("Test RMSE:", test_rmse)
```

```python
#a(repeating the process with  single hidden layers with nodes 5)
mlp=MLPRegressor(hidden_layer_sizes=(5),activation='logistic', solver='lbfgs', random_state=1)

mlp.fit(X_train,y_train)
```

```python
#evaluate model in traing set
y_train_pred=mlp.predict(X_train)
train_rmse=mean_squared_error(y_train,y_train_pred)
print("Training RMSE:", train_rmse)
```

```python
#evaluate model in traing set
y_test_pred=mlp.predict(X_test)
test_rmse=mean_squared_error(y_test,y_test_pred)
print("Test RMSE:", test_rmse)
```

```python
#b(repeating the process with  two hidden layers with nodes 5)
mlp=MLPRegressor(hidden_layer_sizes=(5,5),activation='logistic', solver='lbfgs', random_state=1)

mlp.fit(X_train,y_train)
```

```python
#evaluate model in traing set
y_train_pred=mlp.predict(X_train)
train_rmse=mean_squared_error(y_train,y_train_pred)
print("Training RMSE:", train_rmse)
```

```python
#evaluate model in traing set
y_test_pred=mlp.predict(X_test)
test_rmse=mean_squared_error(y_test,y_test_pred)
print("Test RMSE:", test_rmse)
```

The RMS error for the training data typically goes down as the number of layers and nodes rises.
In  this specific case, we can see that the RMSE for the training data decreased as the numbers layers and node increased.However,In validataion  data,first RMSE is decreased and then second it is increased when nodes and layers both increased.
Hence, it can suggest that the single hidden layer with 2 nodes may be sufficient for this application.
